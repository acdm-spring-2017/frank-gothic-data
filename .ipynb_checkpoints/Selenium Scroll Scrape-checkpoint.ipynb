```python
from selenium import webdriver
path_to_phantomjs = '/Users/lavin/phantomjs-2.1.1-macosx/bin/phantomjs' # change path as needed
browser = webdriver.PhantomJS(executable_path = path_to_phantomjs)
```

```python
url = 'https://www.freshu.io/?section=News'
browser.get(url)
```

```python
html_source = browser.page_source
```

```python
def count_posts(browser):
    from bs4 import BeautifulSoup
    html_source = browser.page_source
    mybs = BeautifulSoup(html_source)
    return len(mybs.find_all("div", {"class":"post-info"}))
```

```python
import time
delay = .75 # seconds

a = 0
while a < 500:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(delay)

    a = count_posts(browser)
    print(a)
```
