```python
%matplotlib inline
#read directory, convert all csv files to dataframes
import glob
import pandas as pd
#isolate the proper column in each
a = glob.glob('*.csv')
#join rows together 
titles = []
for i in a:
    with open(i) as e:
        rows = (e.read().replace("\"", "").split("\n"))
    all_rows = [y.split(",") for y in rows]
    t = [u[0] for u in all_rows]
    titles.extend(t)
    
#join files together
all_tokens = []
for t in titles:
    tokens = t.lower().split(" ")
    all_tokens.extend(tokens)
all_tokens
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud(
                      #font_path='/Users/sebastian/Library/Fonts/CabinSketch-Bold.ttf',
                      stopwords=STOPWORDS,
                      background_color='black',
                      width=1800,
                      height=1400
                     ).generate(' '.join(all_tokens))

plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./my_wordcloud_1.png', dpi=300)
plt.show()
```
